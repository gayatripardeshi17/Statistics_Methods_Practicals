```python
#from a pandas dataframe
batsman = pd.read_csv('/content/matches.csv')
batsman
plt.plot(batsman['season'],batsman['win_by_runs'])
```

```python
#plotting multiple plots
plt.plot(batsman['season'],batsman['win_by_runs'])
plt.plot(batsman['season'],batsman['win_by_wickets'])
```

```python
#color (hex) and line (width and style) and marker (size)
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['season'],batsman['win_by_runs'],color='pink')
plt.plot(batsman['season'],batsman['win_by_wickets'],color='blue')
plt.title('season vs win_by_runs\ wickets')
plt.xlabel('Seasons')
plt.ylabel('Run_wickets_scored')
```

```python
plt.plot(batsman['season'],batsman['win_by_runs'],color='red',linestyle='solid',linewidth=3)
plt.plot(batsman['season'],batsman['win_by_wickets'],color='black',linestyle='solid',linewidth=2)
```

```python
plt.plot(batsman['season'],batsman['win_by_runs'],color='red',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['season'],batsman['win_by_wickets'],color='black',linestyle='solid',linewidth=3,marker='o')
```

```python
plt.plot(batsman['season'],batsman['win_by_runs'],color='red',linestyle='solid',linewidth=1,marker='D',markersize=10,label='Virat')
plt.plot(batsman['season'],batsman['win_by_wickets'],color='black',linestyle='solid',linewidth=1,marker='o',label='Rohit')
plt.legend(loc='upper right')
```
